```python
time.sleep(7)
data = driver.page_source
import bs4
soup = bs4.BeautifulSoup(data)
```

```python
product_name = soup.find_all('div',{'class':'_10Wbs- _2STCsK _3IqNCf'})
```

```python
product_name_list = []
for product in product_name :
    product_name_list.append(product.text)
len(product_name_list)

for i in range(5):
    product_name_list.pop(0)
    product_name_list.pop()
product_name_list #correction check
```

```python
sell_number = soup.find_all('div',{'class':'_1uq9fs'})
```

```python
sell_number_list = []
for sales in sell_number :
    sell_number_list.append(sales.text)


for i in range(5):
    sell_number_list.pop(0)
    sell_number_list.pop()
print(len(sell_number_list)) #correction check
sell_number_list
```

```python
review_number_list = []
```

```python
for i in range(0,50,1):#len(product_name_list)):
    product = driver.find_element(By.XPATH,"//div[.='" + product_name_list[i] + "']")
    driver.execute_script("arguments[0].click();", product)
    driver.execute_script("document.body.style.zoom='30%'")
    print(i)
    print(product_name_list[i])
    time.sleep(5)
    pageData = driver.page_source
    pageSoup = bs4.BeautifulSoup(pageData)
    review_number = pageSoup.find("div", {"class": "product-rating-overview__filter product-rating-overview__filter--with-comment"}).get_text(strip=True)
    print(review_number)
    review_number_list.append(review_number)
    driver.back()
    driver.execute_script("document.body.style.zoom='10%'")
driver.back() #done confirm
```

```python
print(review_number_list)
print(len(review_number_list))
print(len(product_name_list)) #correction check
```
